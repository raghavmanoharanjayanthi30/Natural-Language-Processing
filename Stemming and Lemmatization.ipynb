```python
#stemming with nltk

from nltk.stem import PorterStemmer
#examples of other stemmers are SnowballStemmer
stemmer = PorterStemmer()
words = ["eating", "eats", "eat", "ate", "adjustable", "rafting", "ability", "meeting","better"]
```

```python
for word in words:
    print(word, " | ", stemmer.stem(word))
```

```python
'''
Stemming could be efficient because it is faster than lemmatization, 
does not require language knowledge, fewer rules than lemmatization
'''
```

```python
#lemmatization
nlp = spacy.load("en_core_web_sm")
doc = nlp("eating eats eat ate adjustable rafting ability meeting better")

for token in doc:
    print(token, " | ", token.lemma_, " | ", token.lemma)
```

```python
'''
token.lemma prints unique hash of your lemma which is a base word 
because lemmatization here is a trained english model with a fixed 
vocabulary.it has a few thousand words and for each word it will 
have a unique identifier

'''
```

```python
doc = nlp("Mando talked for 3 hours although talking isn't his thing. he became talkative")

for token in doc:
    print(token, " | ", token.pos_, " | ", token.lemma_, " | ", token.lemma)
```

```python
doc = nlp("Bro, you wanna go? Brah, don't say no! I am exhausted")
#customize lemmatization to convert bro and brah to the base word brother
ar = nlp.get_pipe("attribute_ruler") #retrieves attribute ruler component from the pipeline
ar.add([[{"TEXT": "Bro"}], [{"TEXT": "Brah"}]], {"LEMMA": "Brother"})
```

```python
for token in doc:
    print(token, " | ", token.lemma_)
```
