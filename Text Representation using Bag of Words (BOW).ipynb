```python
import pandas as pd
```

```python
df = pd.read_csv("spam.csv", encoding = "ISO-8859-1") #encoding
#https://stackoverflow.com/questions/18171739/unicodedecodeerror-when-reading-csv-file-in-pandas
```

```python
df = df[['Category', 'Message']]
```

```python
df
```

```python
def get_spam(x):
    if x == 'spam':
        return 1
    return 0

#two ways of doing the same thing
df['spam'] = df.Category.apply(lambda x: 1 if x == 'spam' else 0)
#df['spam'] = df.Category.apply(get_spam)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2)
```

```python
#count vectorizer to build bag of model for our numeric representation
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_cv = v.fit_transform(X_train.values) #.values has to be used because .values is needed to convert series object to numpy array
X_train_cv
```

```python
#import naive bayes classifier

#there are two types of NB classifier
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_cv, y_train)

#evaluate performance of model
X_test_cv = v.transform(X_test.values)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test_cv)
```

```python
print(classification_report(y_test, y_pred))
```

```python
#f1 score is a good metric to use because the training data set is imbalanced without 4800 ham and 750 spam
```

```python
#implement count vectorizer bag of words using sklearn pipeline

from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
